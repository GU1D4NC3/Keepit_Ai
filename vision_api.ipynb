# 이미지 내 음식 객체 검출 모델 코드

```python
import io
import os

# Imports the Google Cloud client library
from google.cloud import vision
# from google.cloud import vision_v1p3beta1 as vision2
# Instantiates a client
client = vision.ImageAnnotatorClient()

# The name of the image file to annotate
file_name = os.path.abspath('c:/Users/hanji/Desktop/test_image/test55.jpg') 
# c:/Users/hanji/Desktop/test_image/test2.jpg - banana
# c:/Users/hanji/Desktop/test_image/test8.jpg - 밥상
# c:/Users/hanji/Desktop/test_image/fruit1.jpg - 과일

# Loads the image into memory
with io.open(file_name, 'rb') as image_file:
    content = image_file.read()

image = vision.Image(content=content)

objects = client.object_localization(image=image).localized_object_annotations

print(f"Number of objects found: {len(objects)}")
for object_ in objects:
    print(f"\n{object_.name} (confidence: {object_.score})")

with open('output.txt', 'w') as txt_file:
    for object_ in objects:
        line = f"{object_.name}\n"
        # line = f"{object_.name} (confidence: {object_.score})\n"
        txt_file.write(line)


# image = vision.Image(content=content)
# # Performs label detection on the image file
        


# response = client.label_detection(image=image)
# labels = response.label_annotations
# print('Labels:')
# for label in labels:
#     print(label.description)




# response = client.image_properties(image=image)
# props = response.image_properties_annotation
# print("Properties:")

# for color in props.dominant_colors.colors:
#     print(f"fraction: {color.pixel_fraction}")
#     print(f"\tr: {color.color.red}")
#     print(f"\tg: {color.color.green}")
#     print(f"\tb: {color.color.blue}")
#     print(f"\ta: {color.color.alpha}")
    
# image = vision.Image(content=content)

# objects = client.object_localization(image=image).localized_object_annotations

# print(f"Number of objects found: {len(objects)}")
# for object_ in objects:
#     print(f"\n{object_.name} (confidence: {object_.score})")
#     print("Normalized bounding polygon vertices: ")
#     for vertex in object_.bounding_poly.normalized_vertices:
#         print(f" - ({vertex.x}, {vertex.y})")
```

# 음식 객체 각각의 영양소 추출 코드

```python
import json

with open('nutrient_info.json', 'r') as f:
    nutrient_dict = json.load(f)

for object_ in objects:
    obj_name = object_.name
    if obj_name in nutrient_dict:
        print(f"\n{obj_name}의 영양소 정보: {nutrient_dict[obj_name]}")
    else:
        print(f"\n{obj_name}에 대한 정보를 찾을 수 없습니다.")
```

# 영양소 총합 계산 코드

```python
total_nutrients = {
    "calories": 0,
    "protein": 0,
    "carbohydrates": 0,
    "fat": 0,
    "fiber": 0,
    "vitamin_C": 0,
    "potassium": 0,
    "calcium": 0,
    "iron": 0
}

for object_ in objects:
    obj_name = object_.name
    if obj_name in nutrient_dict:
        for nutrient, value in nutrient_dict[obj_name].items():
            total_nutrients[nutrient] += value

print("\n모든 객체의 영양소 총합:")
for nutrient, total in total_nutrients.items():
    print(f"{nutrient}: {total}")
```

```python
# 임산부가 하루에 섭취해야하는 영양소 섭취량
daily_intake = {
    "calories": 2200, # 칼로리
    "protein": 70, # 단백질
    "carbohydrates": 175, # 탄수화물
    "fat": 70, # 지방
    "fiber": 28, # 식이섬유
    "vitamin_C": 85, # 비타민C
    "potassium": 4700, # 칼륨
    "calcium": 1000, # 칼슘
    "iron": 27 # 철
    # ... 여기에 필요한 다른 영양소들 추가
}


for nutrient, total in total_nutrients.items():
    if nutrient in daily_intake:
        required = daily_intake[nutrient]
        if total < required:
            print(f"{nutrient}: {required - total}만큼 더 섭취해야 합니다.")
        else:
            print(f"{nutrient}: {total - required}만큼 초과 섭취하였습니다.")
    else:
        print(f"{nutrient}: 이 영양소의 권장 섭취량 정보를 찾을 수 없습니다.")
```
